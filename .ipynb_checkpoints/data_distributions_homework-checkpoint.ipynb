```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
pd.set_option('display.max_rows', 100)
```

```python
# CODE TAKEN FROM THINKFUL 
# Making a standard normally distributed variable with 10000 observations, a mean of 0, and 
# a standard deviation of 1, and putting it in a data frame.

mean = 0 
sd = 1
n = 10000

normal = pd.Series(np.random.normal(mean, sd, n))
# Plotting the variables in the data frame (here, just the variable "rand") as a histogram.
normal.hist(bins=100)
plt.axvline(normal.mean(), color='b', linestyle='solid', linewidth=1)
plt.axvline(normal.mean() - normal.std(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(normal.mean() + normal.std(), color='b', linestyle='dashed', linewidth=1)
# Inline printing the histogram
plt.show()
```

```python
#n = 100 (samples or trials)
#p = 0.5 (probability of success)
# size = 10000 (number of experiments)
binomial = pd.Series(np.random.binomial(100, 0.5, size=100))
binomial.hist(bins=100)
plt.axvline(binomial.mean(), color='b', linestyle='solid', linewidth=1)
plt.axvline(binomial.mean() + binomial.std(), color='b', linestyle='dashed', linewidth=1)
plt.axvline(binomial.mean() - binomial.std(), color='b', linestyle='dashed', linewidth=1)
plt.show()
```

1.Generate two normally-distributed variables, one with a mean of 5 and standard deviation of 0.5, 
  and the other with a mean of 10 and standard deviation of 1.

2.Add them together to create a third variable.

3.Graph the third variable using a histogram.

4.Compute the mean and standard deviation and plot them as vertical lines on the histogram.

5.Evaluate the descriptive statistics against the data.

```python
var1 = np.random.normal(5, 0.5, 100)
var2 = np.random.normal(10, 1, 100)
var3 = var1 + var2

#histogram
fig = plt.figure(figsize=(14,7))
plt.hist(var1, bins=100)
plt.hist(var2, bins=100)
plt.hist(var3, bins=100)

#means
plt.axvline(var1.mean(), color='b', linestyle='solid', linewidth=1)
plt.axvline(var2.mean(), color='b', linestyle='solid', linewidth=1)
plt.axvline(var3.mean(), color='b', linestyle='solid', linewidth=1)

#standard devs
plt.axvline(var1.mean() + var1.std(), color='r', linestyle='dashed', linewidth=1)
plt.axvline(var1.mean() - var1.std(), color='r', linestyle='dashed', linewidth=1)

plt.axvline(var2.mean() + var2.std(), color='r', linestyle='dashed', linewidth=1)
plt.axvline(var2.mean() - var2.std(), color='r', linestyle='dashed', linewidth=1)

plt.axvline(var3.mean() + var3.std(), color='r', linestyle='dashed', linewidth=1)
plt.axvline(var3.mean() - var3.std(), color='r', linestyle='dashed', linewidth=1)

plt.show()
print(var1.mean())
print(var2.mean())
print(var3.mean())
```
